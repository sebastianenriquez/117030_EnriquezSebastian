16 Leer una lista de números ya ordenados de forma ascendente y verificar que dicha lista está ordenada. Luego, leer un número e insertarlo en la lista en la posición que le corresponde a dicho número.

```python
lista=eval(input('Ingrese una lsita de números '))
i=0
t=1
z=0 #funciona para saber si se a cumplido la condición de ordenamiento

while t < len(lista): 
    if lista[i]==lista[t]: #busca que no hayan números repetidos
        print('En la lista hay números repetidos')
        z=1 
        break
    elif lista[i]<lista[t]: #verifica que los números sean ascendentes
        i+=1
        t+=1
    elif lista[i]>lista[t]: #busca si un número es mayor al siguiente
        print('La lista no está ordenada de forma ascendente')
        z=1
        break

if z==0: # si se cumple la condición de orden continúa
    print('La lista se encuentra ordenada de forma ascendente')
    
    x=float(input('Ingrese un número: '))
    k=0 
    while k < len(lista):
        if x==lista[k]: #revisa si el número ya se encuentra en la lista
            print('El número ya se encuentra en la lista')
            z=1
            break
        elif x > lista[k]: #busca donde posicionar el número ingresado
            k+=1
        elif x < lista[k]: #posiciona el número cuando encuentra uno menor
            lista.insert(k,x)
            break
        if x > lista[-1]: #considera el caso que el número ingresado sea mayor
                          #a todos en la lista y lo agrega al final
            lista.append(x)
            break
            
    if z== 0 : print('la nueva lista es:',lista)
```

17 Leer una lista y determinar cuál es el segundo elemento más grande de dicha lista. No utilice ningún tipo de algoritmo de ordenamiento para realizar este ejercicio.

```python
L=eval(input('Ingrese una lista, procure que los elementos de la lista'
             ' sean del mismo tipo: '))

i=0 
may=L[0] #pone como elemento mayor al primero de la lista
while i < len(L):
    if may < L[i]: #compara el elemento mayor hasta el momento con el 
                   #siguiente en la lista, si ese elemto es mayor, entonces 
                   #actualiza la variable may
        may=L[i]
        i+=1
    else: i+=1
    
while may in L: #elimina el elemto mayor de la lista L las veces que 
                #se encuentre
    L.remove(may)

t=0
may1=L[0]
while t < len(L): #vuelve a realizar el mismo procedimiento para buscar el 
                  #mayor de la lista resultante, que es segundo mayor de la 
                  #lista ingresada
    if may1 < L[t]:
        may1=L[t]
        t+=1
    else: t+=1 

print('El segundo elemento más grande de la lista es:',may1)
```

18 Leer una matriz cuadrada A, obtener su matriz triangular superior asociada U y determinar qué fila de U contiene la mayor 
cantidad de unos.

```python
dimen=int(input('Diga la catidad de filas x columna de la matriz: '))

M=[] #crea una lista vacía que almacenará los datos de la matriz
for fila in range(dimen): #se crea una matriz de ceros con la cantidad de filas
                          #y columnas indicadas
    M.append([0]*dimen)

fila=0
col=0
while fila < dimen:
    while col < dimen: #realiza el ingreso de los valores de la matriz
        x=int(input(f'Ingrese el valor de la posición {fila+1},{col+1}: '))
        M[fila][col]=x
        col+=1
    fila+=1
    col=0

print('La matriz que ingresó es: ')
print( )
for fila in M:
    print(fila)
#el siguiente algoritmo realiza la triangularización de la matriz
#aplicando operaciones entre filas, con la teoría de Gauss

z=0 #indica la fila con la que se realiza la operación de suma
t=0 #idica la fila y el elemento de la diagonal que se pretende
    #mulplicar para hacerlo "1"
i=1 #arranca desde la siguinte fila de z para realizar la suma

while z < dimen:
    for fila in M: #organiza la matriz para que las filas de ceros esten al 
                   #final de la matriz y el código no presente problemas
                   #al realizar las operaciones
        if fila == [0]*dimen:
            M.append(fila)
            M.remove(fila)
    if M[t][t]!=0: #verifica que el valor con el que se opera no sea cero
        c=M[t][t] #guarda el valor con el que se pretende hacer "1" el primer
                  #valor no nulo de dicha fila
        for j in range(dimen): #opera los elemtos de la fila con el inverso
                               #del primer valor no nulo 
            M[t][j]=round((M[t][j]*1/(c)),1)
    if M[t][t]==0: # si el valor de la fila con el que se opera es cero
                   #entonces reorganiza la fila hasta encontrar uno diferente
                   #a cero
        p=0
        while t+p < (dimen-1):
            if M[t+p][t]==0: #compara con las filas siguientes buscando una 
                             #fila que tenga un valor diferete a cero
                p+=1
            if M[t+p][t]!=0: # si encuentra la fila mencionanda
                             # hace el cambio de filas
                x=M[t]
                y=M[t+p]
                M[t+p]=x
                M[t]=y
                break
            else: #si no encuentra dicha fila, la deja en su posición y 
                  #continúa con las operaciones entre filas
                M[t]=M[t]
    if M[t][t]!=0: #retoma la operación de hacer 1 el elemento de la diagonal
        c=M[t][t]
        for j in range(dimen):
            M[t][j]=round((M[t][j]*1/(c)),1)
        
    while i+z < dimen: # realiza la operación de suma de filas
                       # para anular los elementos debajo de la diagonal
        b=M[i+z][t] #guarda la variable con la que se opera
        for j in range(dimen):
            M[i+z][j]=round((M[i+z][j] - b*M[t][j]),1)
        i+=1
    t+=1 #cambia al siguiente valor en la diagonal
    z+=1 #cambia a la siguiente fila
    i=1 #reicinia el índice que opera filas

print( )
print('La matriz triangular superior es: ')
print( )
for fila in M:
    print(fila)
# El siguiente se encarga de encontrar la fila con mayor cantidad de unos
may=0 #indica la mayor cantidad de unos
fil=0
ind=0 #indica la fila donde se encuentra la mayor cantidad
while fil < dimen-1:
    u=[]
    v=[]
    #el algoritmo compara dos filas consecutivas
    for j in M[fil]:
        if j==1.0:
            u.append(1)
    for j in M[fil+1]:
        if j==1.0:
            v.append(1)
    if len(u)> len(v): #actualiza la cantidad de may
        if len(u)> may:
            may=len(u)
            ind=fil #identifica en que fila sucedió
    else: 
        if len(v)>may:
            may=len(v)
            ind=fil+1
    
    fil+=1
cont=0 
for h in M[ind]: #verifica que la cantidad de unos que hay en dicha fila
    if h ==1.0:
      cont+=1
      
if cont>1: 
    print('La fila con mayor cantidad de unos es la fila',ind+1)
else: #si se tiene que cont es 1, entonces se verifica si está solo en la 
      #primera fila o si se encuentra en la siguiente
    cont1=0
    for h in M[1]: #como minímo debe haber un uno en la primera fila
                   #de lo contrario se puede concluir
        if h == 1.0:
            cont1+=1
    if cont==0:
        print('En la matriz no hay unos')
    
    elif cont>cont1: #si en la fila 2 no hay unos, entonces se con concluye que
                   #la fila con mayor cantidad de unos es la fila uno
        print('La fila con mayor cantidad de unos es la fila',1)
        
    else: # si en la segunda fila hay un uno, se concluye que hay almenos
          # dos filas con un único uno en ellas
        print('Hay varias filas con un único "1" en ellas')
        
        #estas comparaciones se pueden hacer ya que la matriz fue reducida
        #con operaciones entre filas
```

19 Leer una lista e identifique la sublista más larga dentro de dicha lista que se encuentra ordenada ascendentemente. Por ejemplo, en la lista [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la [1, 2, 6, 9].

```python
L=eval(input('Ingrese una lista de números: '))
U=[] #crea dos listas que se irán actualizando cuando encuentren númmeros 
     #organizados
V=[]
i=0
t=1
while t < len(L)-1:
    if  L[i] < L[t]: #agrega elementos a la lista U, cuando los encuentra 
                     #ordenados
        U.append(L[i])
        i+=1
        t+=1
    
    if L[i] > L[t] and len(U) > len(V): #si la lista U es más grande que la 
                                        #lista V, entonces actualiza V
                                        #esto despues de encontrar un salto
                                        #en el orden
        U.append(L[i])
        V=U[:]
        U=[] #vacía la lista U para que sigan entrando valores
        if t<len(L)-1: 
            i+=1
            t+=1
            
    if L[i] > L[t] and len(U) < len(V): #si encuentra una discontinuidad y la 
                                        #lista U es menor a la lista V, 
                                        #entonces descarta la lista U
        U=[]
        if t<len(L)-1:
            i+=1
            t+=1
    
    if L[i] < L[t] and t==len(L)-1: #considera si los últimos dígitos deben ser
                                    #agregados a la lista U
        U.append(L[i])
        U.append(L[t])
        if len(U) > len(V): #compara cúal es la lista mayor
            
            print('La sublista ordenada más grante es:',U)
        elif len(U)==len(V):
            print('Hay dos sublistas de igual tamaño mayor a las demás,'
                 ' estas son:',U,V)
        else:
            print('La sublista ordenada más grande es:',V)
```

20 Hacer un programa que lea una matriz y la rote 90 grados en sentido antihorario. Imprimir la matriz resultante.

```python
'''Algoritmo realizado sin utilizar la librería numpy'''

n=int(input('Diga la cantidad de filas: '))
m=int(input('Diga la cantidad de columnas: '))

M1=[] #crea una lista vacía que almacenará los datos de la matriz
for fila in range(n): #se crea una matriz de ceros con la cantidad de filas
                      #y columnas indicadas
    M1.append([0]*m)

fila=0
col=0
while fila < n:
    while col < m: #realiza el ingreso de los valores de la matriz
        x=int(input(f'Ingrese el valor de la posición {fila+1},{col+1}: '))
        M1[fila][col]=x
        col+=1
    fila+=1
    col=0

print('La matriz que ingresó es: ')
print( )
for fila in M1:
    print(fila)
    

M2=[] #crea una lista vacía que almacenará los datos de la matriz rotada
for fila in range(m): #se crea una matriz de ceros con la cantidad de filas
                      #y columnas indicadas
    M2.append([0]*n)


#variables que indican el índice para cambiar los valores
fil1=0 #indica la posición de fila para M1
fil2=m-1 #indica la posición fila de M2
col1=0 #posición columna de M1
col2=0 #posición columna de M2
while col2 >= 0: #condiciona el ciclo infinitamente
    if fil1 == n or col1 == m: #rompe el ciclo cuando se ha recorrido toda la
                               #matriz ingresada
        break
    while fil2 >= 0: #impide que el valor que desciende sea negativo
        M2[fil2][col2]=M1[fil1][col1] #cambia los valores de M2 en su posición 
                                      #correspondiente
        fil2-=1
        col1+=1

    fil1+=1
    col2+=1
    col1=0 #reinicia la posición columna que varía en el ciclo interno
    fil2=m-1 #reinicia la posición de fil2 para el ciclo interno
            
print('La matriz rotada 90° en sentido antihorario es:')
print( )

for fila in M2:
    print(fila)
```
